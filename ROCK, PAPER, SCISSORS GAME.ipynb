```python
import random
l=["rock", "scissor", "paper"]

while True:
    ccount=0;
    ucount=0;
    
    Name=input("Enter your name here: ")
    print("""
Winning Rules:
1. Paper vs Rock --> Paper
2. Rock vs Scissors --> Rock
3. Scissors vs Paper --> Scissors
""")     
    

    uc=int(input('''   
Game Start.....
1 Yes
2 No | Exit
                 '''))

    if uc==1:
        for a in range(1,6):
            userInput=int(input('''
            Enter your choice from 1-3:
1 Rock
2 Scissor
3 Paper
'''))
            if userInput==1:
                uchoice="rock"
            elif userInput==2:
                uchoice="scissor"
            elif userInput==3:
                uchoice="paper"
            Cchoice=random.choice(l)
            
            if Cchoice==uchoice:
                print("Computer Value",Cchoice)
                print("User Value",uchoice)
                print("Game Tie")
                ucount=ucount+1;
                ccount=ccount+1;
            elif(uchoice=="rock" and Cchoice=="scissor") or (uchoice=="paper" and Cchoice=="rock") or (uchoice=="scissor" and Cchoice=="paper"):
                print("Computer Value",Cchoice)
                print("User Value",uchoice)
                print("You Win")
                ucount=ucount+1;
            else:
                print("Computer Value",Cchoice)
                print("User Value",uchoice)
                print("You lose")
                print("Computer Win")
                ccount=ccount+1;   
                
        if ucount==ccount:
            print("\nFinal Game Tie....")
            print(Name,"'s Score is",ucount)
            print("Computer Score",ccount)
        elif ucount > ccount:
            print("\nFinal You Win The Game....")
            print(Name,"'s Score is",ucount)
            print("Computer Score",ccount)
        else:
            print("\nFinal Computer Win The Game....")
            print(Name,"'s Score is",ucount)
            print("Computer Score",ccount)

            print("Game Over!")
            print("Thanks for Playing")
    else:
        break;
```
